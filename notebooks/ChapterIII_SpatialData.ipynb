## Types of Spatial Data
Three principal types of spatial data
### Areal Data
A spatial plane is partitioned into a finite number of areal units, or polygons. In the case of regular areal data, the polygon takes the form of regular grid of square or rectangular object while generally the polygons are irregular and differ in shape from unit to unit (Eg. block groups, census tracts, zip codes, counties, and countries). 
### Geostatistical Data
When observed data are sample data from a continuos underlying surface. Eg includes fertility rates, politcal campaign contributions, and housing prices that are observed at particular sample locations and unobserved at other locations.
### Point Pattern Data
When observed spatial locations are the locations of discrete events.

## Representing Spatial Data in Computers
### Vector (Object View)
Represent finite set of entities (shapes/geometries). Include Polygon, Lines, and Points

```python
#Reading a shapefile using geopandas in Python
import geopandas as gpd
oh_counties=gpd.read_file(r'../class_data/Ohio_counties/ODOT_County_Boundaries.shp')
#show attribute values for first five records
#print (oh_counties.head())
#plot oh_counties
oh_counties.plot()
```

### Raster (Field View) 
Images encoding surfaces (values, colors etc.). Represent continuos data such as elevation.

```python
#Reading a raster GeoTIFF using rasterio in Python
import rasterio
from rasterio.plot import show
src=rasterio.open(r'../class_data/PM_25_CA.tif')
show(src.read(), transform=src.transform, cmap='viridis')
```

## Traditional and New Sources of Spatial Data
### Traditional Sources (Good Old Spatial Data)
Eg. Census Data

<ul>

<li><b>Detailed</b> Information</li>

</ul>
